```python
# Import các thư viện thông dụng
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import các distribution packages từ thư viện scipy
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm
from scipy.stats import t
```

```python
# 1 Đọc dữ liệu và hiển thị 10 dòng đầu tiên của dữ liệu
df = pd.read_csv('datasets/bds.csv', header=0)

df
```

```python
# 2a Số lượng loại mỗi loại (type) bất động sản có trong dữ liệu
df.groupby('type').size().reset_index(name='Số lượng').rename(columns={'type': 'Loại BDS'})
```

```python
# 2b  Số lượng bất động sản của mỗi loại (type) theo từng cách thức rao bán (poster_temp).
result = df.pivot_table(index='type', columns='poster_temp', aggfunc='size', fill_value=0).reset_index()

result
```

```python
plt.subplot(1,2,1)
plt.pie(result['khác'], labels=result['type'], autopct='%1.1f%%', startangle=90)
plt.title('% các loại BDS theo hình thức rao bán khác')
plt.plot()

plt.subplot(1,2,2)
plt.pie(result['môi giới'], labels=result['type'], autopct='%1.1f%%', startangle=90)
plt.title('% các loại BDS theo hình thức rao bán môi giới', loc='center', y=-0.1)
plt.plot()
```

```python
# Chuyển dữ liệu time thành kiểu datetime, vẽ biểu đồ thể hiện số lượng bất động sản được bán theo từng tháng trong năm 2020 và đưa ra nhận xét
df['time'] = pd.to_datetime(df['time'])

df
```

```python
df_2020 = df[df['time'].dt.year == 2020].copy()
df_2020['month'] = df_2020['time'].dt.month

result = df_2020.groupby('month').size()

plt.bar(result.index.tolist(), result.tolist())
plt.title('số lượng bất động sản được bán theo từng tháng trong năm 2020')
plt.show()
```
