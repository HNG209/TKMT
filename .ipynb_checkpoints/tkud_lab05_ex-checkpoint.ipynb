**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy:

**Bài 04:** Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.

```python
num_rolls = 5000

# Mô phỏng tung con xúc sắc 5000 lần
rolls = np.random.randint(1, 7, num_rolls)  # Tạo ra 5000 số ngẫu nhiên từ 1 đến 6
roll_gt4 = rolls[np.where(rolls >= 4)]

# Đếm số lần xuất hiện số 4
count_4 = np.sum(rolls == 4)
count_gt4 = np.sum(rolls >= 4)
count_6 = np.sum(roll_gt4 == 6)

# Tính xác suất xuất hiện số 4
prob_4 = count_4 / num_rolls
prob_gt4 = count_gt4 / num_rolls
prob_6 = count_6 / roll_gt4.size

print("Xác suất để số chấm xuất hiện là 4:", prob_4)
print("Xác suất để số chấm xuất hiện >= 4:", prob_gt4)
print("xác suất để mặt 6 chấm xuất hiện:", prob_6)
```

Câu lệnh trên cho biết xác suất có đúng 4 lần thành công trong 10 phép thử, khi xác suất thành công mỗi lần là 60% (0.6).

b. Giả sử X có phân phối nhị thức `X ~ B(10, 0.6)`. Hãy tính các giá trị sau:  
$P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

c. Tính các giá trị sau và vẽ hình minh họa cho các giá trị tính được:

- `norm.ppf(0.975)`

- `norm.ppf(0.975, 1, 1)`

- `norm.ppf(0.975, 1, 2)`

```python
# Tính các giá trị ppf
ppf1 = norm.ppf(0.975)
ppf2 = norm.ppf(0.975, 1, 1)
ppf3 = norm.ppf(0.975, 1, 2)

# In ra kết quả
print("norm.ppf(0.975) =", ppf1)
print("norm.ppf(0.975, 1, 1) =", ppf2)
print("norm.ppf(0.975, 1, 2) =", ppf3)

# Tạo khoảng giá trị x cho biểu đồ
x = np.linspace(-5, 8, 1000)

# Tính mật độ xác suất (PDF) cho các phân phối
pdf1 = norm.pdf(x, 0, 1)  # Phân phối chuẩn tắc
pdf2 = norm.pdf(x, 1, 1)  # Phân phối chuẩn với mean=1, std=1
pdf3 = norm.pdf(x, 1, 2)  # Phân phối chuẩn với mean=1, std=2

# Vẽ đồ thị
plt.figure(figsize=(10, 6))

# Đường cong phân phối chuẩn tắc
plt.plot(x, pdf1, label='N(0, 1)', color='blue')
plt.axvline(x=ppf1, color='blue', linestyle='--', label=f'ppf(0.975) ≈ {ppf1:.2f}')

# Đường cong phân phối với mean=1, std=1
plt.plot(x, pdf2, label='N(1, 1)', color='green')
plt.axvline(x=ppf2, color='green', linestyle='--', label=f'ppf(0.975, 1, 1) ≈ {ppf2:.2f}')

# Đường cong phân phối với mean=1, std=2
plt.plot(x, pdf3, label='N(1, 2)', color='red')
plt.axvline(x=ppf3, color='red', linestyle='--', label=f'ppf(0.975, 1, 2) ≈ {ppf3:.2f}')

# Thiết lập đồ thị
plt.title("Biểu đồ phân phối chuẩn với các giá trị ppf(0.975)")
plt.xlabel("Giá trị")
plt.ylabel("Mật độ xác suất (PDF)")
plt.legend()
plt.grid(True)

# Hiển thị đồ thị
plt.show()
```

d. Giả sử biến ngẫu nhiên X có phân phối chuẩn với kỳ vọng là 24 và phương sai là 16. Hãy tính các giá trị sau:

**Bài 9**

Giả sử điểm thi TOEIC là một biến ngẫu nhiên có phân phối chuẩn với kỳ vọng là 500, độ lệch chuẩn là 8. Bạn hãy:

a. Mô phỏng bằng cách phát sinh ngẫu nhiên điểm thi TOIEC của một nhóm gồm 50 người.

b. Tính điểm trung bình và độ lệch chuẩn về điểm của nhóm.

c. Tìm miền giá trị, và miền phân vị (IQR) của nhóm

d. Cho biết tỷ lệ đạt trên 450 điểm của nhóm. 

e. Vẽ đồ thị histogram của nhóm và so sánh với đồ thị hàm mật độ của phân phối chuẩn với kỳ vọng là 500 và độ lệch chuẩn là 8.

f. Thay vì mô phỏng 50 người, hãy mô phỏng nhóm với 100 người thi. Bạn có nhận xét gì?

```python
mu = 500
sigma = 8

scores = np.random.normal(mu, sigma, 50)

print(scores)

print("Điểm trung bình:", scores.mean())
print("Độ lệch chuẩn:", scores.std())

print("Miền giá trị:", scores.max() - scores.min())
print("Miền phân vị(IQR):", np.quantile(scores, 0.75) - np.quantile(scores, 0.25))

prob_gt450 = scores[np.where(scores > 450)]

print(prob_gt450.size)

print("tỷ lệ đạt trên 450 điểm của nhóm:", prob_gt450.size / scores.size)

print("mô phỏng 100 người")

scores = np.random.normal(mu, sigma, 100)

print("Điểm trung bình:", scores.mean())
print("Độ lệch chuẩn:", scores.std())

print("Miền giá trị:", scores.max() - scores.min())
print("Miền phân vị(IQR):", np.quantile(scores, 0.75) - np.quantile(scores, 0.25))

prob_gt450 = scores[np.where(scores > 450)]

print(prob_gt450.size)

print("tỷ lệ đạt trên 450 điểm của nhóm:", prob_gt450.size / scores.size)
```

**Bài 10**

Viết một chương trình để mô phỏng việc tung đồng xu và đánh giá phương sai của giá trị mô phỏng của biến ngẫu nhiên X (tung được mặt `Head`) đối với các số lần tung khác nhau. Giả sử  đồng xu phải cân bằng. 

Đối với mỗi N từ 10; 40; 90; 160; 250; 490; 640; 810; 1000, hãy ước tính giá trị của X bằng cách mô phỏng số lần tung đó. 

Bạn nên chạy mỗi mô phỏng 100 lần và sử dụng tập hợp các ước tính để đánh giá phương sai của ước tính x của bạn. Vẽ biểu đồ phương sai này so với N, bạn có nhận xét gì không?

```python
N = [10, 40, 90, 160, 250, 490, 640, 810, 1000]
num_simulations = 100

variances = []

for i in N:
    estimates = [np.mean(np.random.binomial(1, 0.5, i)) for _ in range(num_simulations)]
    variance = np.var(estimates)
    variances.append(variance)

plt.plot(N, variances)
```

**Nhận xét**

Phương sai của ước tính X giảm dần khi số lần tung 𝑁 tăng lên.
Phương sai giảm mạnh ở những giá trị N nhỏ, sau đó giảm chậm dần khi N lớn.
Biểu đồ có dạng giảm dần và tiệm cận về 0, cho thấy rằng với số lần tung rất lớn, phương sai gần như bằng 0.
